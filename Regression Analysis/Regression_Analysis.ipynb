```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
```

```python
df = pd.read_csv('/content/Advertising.csv')
df.head()
```

```python
df.columns
```

```python
X = df[['TV','Radio','Newspaper','Sales']]
X = sm.add_constant(X)
```

```python
y = df['Sales']
```

```python
model = sm.OLS(y,X).fit()
```

```python
print(model.summary())
```
